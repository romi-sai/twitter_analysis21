```python
import matplotlib.pyplot as plt
import datetime
#from datetime import datetime
from csv import reader
import tweepy
import time
import csv
import os
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

consumer_key = ""
consumer_secret = ""
access_token = ""
access_token_secret = ""

auth = tweepy.AppAuthHandler(consumer_key, consumer_secret)
api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
```

```python
userLookupList = []
multiple_sn= []
filenames= []

multiple_sn= ["_kenny_joseph"]
filenames=["_kenny_joseph.csv"]
length = len(filenames)

def new_bounds(input_list, file):
    id_list=[]
    name_list=[]
    date_list=[]
    lb_list=[]
    lbub_list=[]
    ubub_list=[]
    users =[]

    fields = ['Screen Name','Creation Date', 'Lower Bound', 'Lower bound of Upper bound','Upper bound of Upper Bound']


    for i in range(0,len(input_list),100):
        users += api.lookup_users(input_list[i:i+100])

    with open(file,'w') as csvfile:
        new_writer = csv.writer(csvfile) 
        new_writer.writerow(fields)
        
        for user in users:
            createdAt = user.created_at.date()
            date_list.append(createdAt)

        for user in users:
            userID = user.id
            id_list.append(userID)

        for user in users:
            screenName = user.screen_name
            name_list.append(screenName)

        i=0 #lowerbound calculations
        string = "Day of API request" #upperbound calculations

        for i in range(len(date_list)-1):
            curr = date_list[i]
            for j in range(i+1, len(date_list)):
                if(date_list[j] > curr):
                    curr = date_list[j]

            lb_list.append(curr)
            ubub_list.append(string)


        j = 0
        for j in range(len(date_list)):
            if(j == (len(date_list)-1)):
                lb_list.append(date_list[j])
                ubub_list.append(string)

        lbub_list.append(string)

        for k in range(1,len(date_list)) :
            current = date_list[k]
            maximum = 0
            for l in range(k-1,-1,-1):
                if(date_list[l] > current):
                    maximum = date_list[l]
                    lbub_list.append(date_list[l])
                    break
            if(maximum == 0):
                lbub_list.append(string)

        #date_list.reverse()
        #name_list.reverse()
        #lb_list.reverse()
        #lbub_list.reverse()
        #ubub_list.reverse()




        for m in range(len(id_list)):
            new_writer.writerow([str(id_list[m]), str(name_list[m]), str(date_list[m]), str(lb_list[m]), str(lbub_list[m]), str(ubub_list[m])])


def friends_count(id):
    user = api.get_user(id)
    followers_count = user.friends_count
    return followers_count
  




#STEP1: Collect who everyone follows and write out to individual files
def get_friends(name, filename, is_screenname):

    total_users = 0
    friend_list = []
    #print("SAVING: {}".format(filename))
    #if the file already exists: 
    try: 
      #  if os.path.exists(filename):
      #  return [x.strip() for x in open(filename)]
      #  else:
        if filename is None: 
            if is_screenname:
                for friend in tweepy.Cursor(api.friends_ids, screen_name= name, count = 5000).items():
                    # total_users+=1
                    friend_list.append(friend)

            else:
                for friend in tweepy.Cursor(api.friends_ids, user_id = name, count = 5000).items():
                    #  total_users+=1
                    friend_list.append(friend)

        else: 
            with open(filename, 'w') as new: 
                if is_screenname:
                    for friend in tweepy.Cursor(api.friends_ids, screen_name= name, count = 5000).items():
                        # total_users+=1
                        friend_list.append(friend)
                        new.write(str(friend)+'\n')

                else:
                    for friend in tweepy.Cursor(api.friends_ids, user_id = name, count = 5000).items():
                        #  total_users+=1
                        friend_list.append(friend)
                        new.write(str(friend)+'\n')
         
        return friend_list
    except tweepy.TweepError as ex: 
    # we get a tweep error when we can't view a user - skip them and move onto the next.
    # don't increment i as we want to replace this user with someone else. 
        if ex.reason == "Not authorized.":
            print("Could not view user {}, skipping...".format(name)) 
    
    except tweepy.TweepError:
        time.sleep(60)
    
    # break


```

```python
newnewfiles=[]
newnewfriends=[]
newfriends_list=[]
#final_friends=[]

newFriends = get_friends(multiple_sn[0], filenames[0], True)
newfriends_list.append(newFriends)

for i in range(len(newFriends)):
    newnewfiles.append(str(newFriends[i]) + '.csv')
    
for j in range(len(newFriends)):
    anotherFriends = get_friends(newFriends[j], newnewfiles[j], False)
    newnewfriends.append(anotherFriends)  

for k in range(len(newnewfriends)):
    if(newnewfriends[k] != None):
        newbounds = new_bounds(newnewfriends[k], newnewfiles[k])
        #final_friends.append(newfriends_list[k])
```

# Code to extract x and y arrays for the scatterplot

```python
def extract_lb(filename):
    store_list=[]
    only_lb=[]
  #  print('hello')
    #for l in range(len(newnewfiles)):
    with open(filename) as newnew:
       # print('hi')
        readCSV = csv.reader(newnew, delimiter=',')
        for row in readCSV:
            store_list.append(row)
         #   print(row)
        
        store_list.pop(0)

        for m in range(len(store_list)):
        #print(store_list[m][3])
            only_lb.append(store_list[m][3])

        return only_lb

#extract_lb(newnewfiles[1])
```

```python
l=0
y_array=[]
x_array=[]
target = '2020-01-01'

for n in range(len(newnewfriends)):
    if(newnewfriends[n] != None):
        count = friends_count(newfriends_list[0][n])
        x_array.append(count)

#print(x_array)
for l in range(len(newnewfriends)):
    lb_count = 0
    if(newnewfriends[l] != None):
        #print(newnewfiles[i])
        new_lbs = extract_lb(newnewfiles[l])
        for m in range(len(new_lbs)):
            if(new_lbs[m] >= target):
                lb_count+=1

        y_array.append(lb_count)
    
print('x_array = ' + str(x_array))
print('y_array = ' + str(y_array))

###SCATTERPLOT
import matplotlib.pyplot as plt
plt.xlabel("Total number of followees")
plt.ylabel("Number of ppl whose lb is >= 2020")
plt.scatter(x_array, y_array)
plt.show()
```
